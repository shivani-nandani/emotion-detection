```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_raw = cv2.imread('test_images/test_image.jpg')
```

```python
type(img_raw)
```

```python
img_raw.shape
```

```python
test_img = cv2.cvtColor(img_raw, cv2.COLOR_BGR2RGB)
plt.imshow(test_img)
plt.show()
```

```python
test_img_gray = cv2.cvtColor(img_raw, cv2.COLOR_BGR2GRAY)
plt.imshow(test_img_gray, cmap='gray')
plt.show()
```

```python
# loading the classifier 
haar_cascade_face = cv2.CascadeClassifier('F:/opencv/build/etc/haarcascades/haarcascade_frontalface_default.xml')
```

```python
# detect face
faces_rects = haar_cascade_face.detectMultiScale(test_img_gray, scaleFactor = 1.2, minNeighbors = 5)
faces_rects
```

```python
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_img, (x, y), (x+w, y+h), (0, 255, 0), 2)
```
